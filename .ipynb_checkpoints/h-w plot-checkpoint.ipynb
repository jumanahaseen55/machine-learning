```python
import sklearn
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
 
```

Data analysing

```python
data={"height":[160,170,162,182,177],"weight":[55,65,45,74,85]}
```

```python
a=pd.DataFrame(data)
```

```python
a
```

```python
plt.scatter(a["height"],a["weight"])
```

Data preprocessing

```python
x=a[["height"]]
y=a[["weight"]]
```

Model creation

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

Training

```python
model.fit(x,y)
```

Prediction/testing

```python
model.predict([[170]])
```
